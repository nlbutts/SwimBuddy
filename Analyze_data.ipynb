# Elly Swim Data
This notebook analyzes data that Elly took swwimming.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)

x = pd.read_csv("data/SensorTile_Log_N005.csv", header=1)
y = np.array(x)
t = y[:,0]
t = t - t[0]
accel = y[:,1:4]
gyro = y[:,4:7]
mag = y[:,7:10]
pressure = y[:,10]
plt.plot(t, accel)
plt.grid()
plt.legend(['X','Y','Z'])
plt.xlabel('Time [ms]')
plt.ylabel('mg')
```

```python
plt.plot(t, gyro)
plt.grid()
plt.legend(['X','Y','Z'])
plt.xlabel('Time [ms]')
plt.ylabel('mdps')
```

```python
plt.plot(t, mag)
plt.grid()
plt.legend(['X','Y','Z'])
plt.xlabel('Time [ms]')
plt.ylabel('mgauss')
```

```python
plt.plot(t, pressure)
```
